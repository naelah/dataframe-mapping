```python
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('postcode.csv')
df
```

```python
postcode_data = pd.read_csv('malaysia_zipcode.csv')
postcode_data.head()
```

```python
def postcode_to_state(postcode):

   state = postcode_data['state'][postcode_data.index[(postcode_data['code_start'] <= int(postcode)) & (postcode_data['code_end'] >= int(postcode)) ].tolist()[0]] 
   return state if state else np.nan
```

```python
#creating new column state based on postcode column
df['state'] = df.postcode.apply(lambda x: postcode_to_state(x)  if(x is not None) else x)
df
```
